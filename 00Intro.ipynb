```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2, 10)
x
```

```python
plt.plot(x, 'o-', label='linear' )
plt.legend()
```

```python
plt.plot(x, x, 'o-', label='linear')
plt.plot(x, x**2, 'x-', label='quadratic')
plt.legend(loc = 'best')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
```

```python
samples1 = np.random.normal(size = 10000,loc=1,scale=0.5)
samples2 = np.random.standard_t(df=10,size=10000)
```

```python
binds = np.linspace(-3,3,50)
plt.hist(samples1, bins=binds, alpha=0.5, label='samples 1')
plt.hist(samples2, bins=binds, alpha=0.5, label='samples 2')
plt.legend(loc='upper left')
```

```python
plt.scatter(samples1, samples2, alpha=0.1)
```
